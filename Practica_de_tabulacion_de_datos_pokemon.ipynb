```python
plt.boxplot(df_pokemon["Attack"])
plt.title("Distribucion de Ataque")
plt.xlabel("Attack")
plt.ylabel("Attack points")
plt.show()
```

```python
fig =plt.figure(figsize=(5, 5))
plt.hist(df_pokemon["Attack"])
plt.title("Distribucion de Ataque")
plt.xlabel("Attack")
plt.ylabel("Attack points")
plt.show()
```

```python
sns.boxplot(data=df_pokemon, x="Attack")
plt.show()
```

```python
df_pokemon.columns
```

```python
fig =plt.figure(figsize=(10, 10))
sns.boxplot(data=df_pokemon, y="Attack", x="Type 2")
plt.title("Comparacion de ataque Type 2")
plt.xticks(rotation=45)
plt.show()
```

```python
fig =plt.figure(figsize=(10, 10))
sns.boxplot(data=df_pokemon, y="Attack", x="Type 1", hue="Legendary")
plt.title("Comparacion de ataque Type 1 legendarios")
plt.xticks(rotation=45)
plt.show()
```

```python
fig =plt.figure(figsize=(7,7))
sns.scatterplot(data=df_pokemon, y="Attack", x="Defense", hue="Legendary")
plt.title("Comparacion de ataque y defensa pkm legendarios")
plt.xticks(rotation=45)
plt.show()
```
